### fine_dust.xlsx 엑셀 파일 불러오기, 라이브러리 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

data = pd.read_excel('fine_dust.xlsx') #인덱스 행이 기본으로 나온다. 0 1 2 3...
data
```

## 2018년 지역별 미세먼지 선 그래프

```python
#data = pd.read_excel('fine_dust.xlsx', index_col = 'area') --> 컬럼을  area로 설정
data2018 = data[2018]  # 데이터를 시리즈로 추출
data2018
```

```python
plt.figure(figsize=(15, 4))                 # 가로 15인치, 세로 4인치 도화지를 만듦.
plt.plot(data2018, color='b',marker='o' )   # 시리즈형으로  data가 들어가 있음.
plt.title('2018년 미세먼지 선 그래프')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.grid()        # 격자무늬
plt.show()
```

```python
plt.figure(figsize=(15, 4))
for year in range(2016, 2020):
    chartdata = data[year]
    plt.plot(chartdata, marker='v', label=year)  # marker --> 's'는 사각형, 'v' 역삼각형, '^' 삼각형


plt.title('2016년~2019년 미세먼지 선 그래프')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.legend()
plt.grid()      
plt.show()
```

## 2017년 지역별 미세먼지 셀 막대 그래프

- plt.bar(x=x의 데이터, height=y의 데이터) # bar -->막대그래프

```python
data2017 = data[2017]

plt.figure(figsize=(15, 4))
plt.bar(data2017.index, data2017, color='g')
plt.title('2017 Fine Dust Bar Graph')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.ylim(35, 55) 
plt.grid(axis='y') 
plt.show()
```

```python
import numpy as np
index =np.arange(4) # 0 ~ 3

plt.figure(figsize=(10, 4))
for year in range(2016, 2020):
    chartdata = data4[year]
    plt.bar(index, chartdata, width=0.2, label=year)
    index = index + 0.2

plt.title('2016~2019 미세먼지 그룹(경기, 인천, 부산, 대구)별 세로 막대 그래프')
plt.ylim(35, 55) 
plt.xlabel('area')
plt.xticks(index-0.5, ['경기', '인천', '부산', '대구'])
plt.ylabel('micrometer')

plt.show()
```
