```python
cafe_menu = [
    {"name": "아메리카노", "price": 4000, "category": "커피"},
    {"name": "카페라떼", "price": 4500, "category": "커피"},
    {"name": "바닐라라떼", "price": 5000, "category": "커피"},
    {"name": "녹차라떼", "price": 5200, "category": "라떼"},
    {"name": "딸기스무디", "price": 5800, "category": "스무디"},
    {"name": "망고에이드", "price": 5500, "category": "에이드"},
    {"name": "치즈케이크", "price": 6000, "category": "디저트"},
    {"name": "초코쿠키", "price": 2500, "category": "디저트"},
]
```

```python
print("☕ 스마트 카페 메뉴 조회 시스템에 오신 것을 환영합니다! ☕")
print("=" * 50)
print("1: 가격별 조회 | 2: 종류별 조회 | 3: 메뉴명 검색 (주문)")
print("=" * 50)
```

```python
# 1. 카페 메뉴 데이터 초기화 (리스트 내 딕셔너리 구조)



# 사용자에게 원하는 기능 입력받기
choice = input("원하는 기능의 번호를 입력하세요 (1, 2, 3): ").strip()

# --- [요구사항 1] 가격 범위별 조회 ---
if choice == "1":
    print("\n--- [가격별 조회] ---")
    try:
        min_price = int(input("최저 가격을 입력하세요 (예: 4000): "))
        max_price = int(input("최고 가격을 입력하세요 (예: 5500): "))
        
        print(\n📊 {min_price}원 ~ {max_price}원 사이의 메뉴 목록:")
        found = False
        for menu in cafe_menu:
            if min_price <= menu["price"] <= max_price:
                print(f"- {menu['name']} ({menu['price']}원) [{menu['category']}]")
                found = True
        
        if not found:
            print("❌ 해당 가격대의 메뉴가 존재하지 않습니다.")
    except ValueError:
        print("❌ 올바른 숫자 형태로 가격을 입력해주세요.")

# --- [요구사항 2] 종류(카테고리)별 조회 ---
elif choice == "2":
    print("\n--- [종류별 조회] ---")
    # 중복 없이 현재 등록된 카테고리 보여주기
    categories = list(set(menu["category"] for menu in cafe_menu))
    print(f"현재 선택 가능한 종류: {', '.join(categories)}")
    
    target_category = input("조회할 메뉴 종류를 입력하세요: ").strip()
    
    print(\n📂 [{target_category}] 종류의 메뉴 목록:")
    found = False
    for menu in cafe_menu:
        if menu["category"] == target_category:
            print(f"- {menu['name']} ({menu['price']}원)")
            found = True
            
    if not found:
        print(f"❌ '{target_category}' 종류의 메뉴는 없습니다.")

# --- [요구사항 3] 메뉴명 검색 및 상세 정보 출력 ---
elif choice == "3":
    print("\n--- [메뉴명 검색 및 주문] ---")
    search_name = input("주문을 원하는 메뉴 이름을 입력하세요: ").strip()
    
    found = False
    for menu in cafe_menu:
        if menu["name"] == search_name:
            print("\n🔍 검색 및 주문 내역 확인:")
            print(f"📌 이름: {menu['name']}")
            print(f"💵 가격: {menu['price']}원")
            print(f"🏷️ 종류: {menu['category']}")
            found = True
            break # 원하는 메뉴를 찾았으므로 반복문 종료
            
    if not found:
        print(f"❌ 입력하신 '{search_name}' 메뉴는 존재하지 않습니다.")

else:
    print("❌ 잘못된 번호를 선택하셨습니다. 프로그램을 종료합니다.")
```
