# Matplotlib 라이브러리 기본 시각화

```python
# 라이브러리 불러오기

import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기

xdata = [2, 3, 4, 5]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata)    #선그래프(꺽은선 그래프, 라인 그래프)
plt.title('Line Graph')   # 차트 제목 
plt.legend() # 범례 
plt.show()
```

# 범례 --> plt.legend()

```python
data1 = [2, 3, 4, 5]
data2 = [8, 6, 4, 2]
plt.plot(data1, label = 'asc')    # 오름차순 data1을 y축 값으로 사용. x 축은 자동생성
plt.plot(data2, label = 'desc')   # 내림차순 data2을 y축 값으로 사용. x 축은 자동생성
plt.title('Line Graph')  
plt.legend()                      # 범례 
plt.show()
```

# 선에 색깔 넣기

```python
xdata = [2, 3, 4, 5]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='red')    # color 선 색깔  바꾸기
plt.title('Line Graph')   # 차트 제목 
plt.legend() # 범례 
plt.show()
```

## x축 제목, y축 제목
- plt.xlabel()
- plt. ylabe()

```python

xdata = [2, 3, 4, 5]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color ='c') # 'c'는 청록색
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.title('Line Graph') 
plt.show()
```

## 선 종류

- linestyle = 'dash' 또는 linestyle = '--'
- 일반 실선 limestyle = 'solid' 또는 linestyle = '-'

```python
plt.plot(data1, color ='r', label = 'dashed', linestyle= '--')   #  대시 긴 점선 --> 파선(쇄선)
plt.plot(data2, color ='g', label = 'dotted', linestyle= ':')    #  점선
plt.plot(ydata, color ='b', label = 'Dash-dot', linestyle= '-.') # 1점 쇄선
plt.legend()  # 범례는 labe에 입력된 내용이 출력
plt.show()
```

# Marker (마커) --> 점 모양

```python
plt.title('X, Y range') 
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color ='b', linestyle = '--', marker = 'o', markersize = '10')
plt.show()
```

```python
plt.title('X, Y range') 
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color ='b', linestyle='--', marker='o', markersize='10')
plt.xlim(0, 25)
plt.ylim(-2, 10)
plt.show()
```
