# 클래스(Class)와 객체 지향 프로그래밍(OOP)

## 1. 클래스란 무엇인가??

**클래스(class)** 는  데이터와 그 데이터를 다루는 함수(메서드)를 하나로 묶어 새로운 타입을 정의하는 방법(==설계도)

| 용어 | 설명 | 비유 |
|------|------|-----|
| 클래스(Class) | 설계도 | 붕어빵 틀, 와플 틀, 과자 틀 |
| 인스턴스(instance) | 클래스로 만든 실제 객체 | 붕어빵 1개개, 와플 1개, 과자 1개 |
| 메서드(method) | 클래스 내부에서 만들어진 함수 | 붕어빵의 기능, 와플의 기능, 과자의 기능 |
| 속성(attribute) |인스턴스가 가진 데이터(인스턴스 변수) |붕어빵의 재료(맛), 캐릭터의 머리 색깔 등..|

# 2. 클래스(Class) 기본 구조

```python
# 클래스 정의 기본 구조
# Class 클래스이름: 으로 시작한다.
# 파이썬에서 클래스이름은 첫글자를 대문자로 하는 것이 관행이다.
class Toy:


    # 메서드(함수) 정의
    # __**__ : 스페셜 메서드라 부름.

    # __init__: 객체(인스턴스)가 처음만들어질 때 자동으로 호출되는 '초기화 메서드'(생성자)
    # self : 현재 만들어진 인스턴스(객체) 자기 자신을 가리키다. 인스턴스 메서드. 인스턴스 변수
    def __init__(self):    
        self.elems =[]  #self.속성 --> 인스턴스 변수 : 각 객체가 개별로 가지는 데이터

    #메서트(함수) 정의 --> 인스턴스 메서드 정의
    def add(self, new_elems):
        """
        첫번째 매개변수는 반드시 self --> 호출한 인스턴스가 자동으로 전달된다.
        new_elems 매개변수는 리스트이다.
        """
        self.elems += new_elems # new_elema 만큼 누적해라!

    #메서트(함수) 정의 --> 인스턴스 메서드 정의
    def size(self):
        """
        저장된 요소 개수를 반환(리턴)한다.
        개수를 우리에게 알려준다.
        """
        return len(self.elems)
    
```

## 3. 클래스로 인스턴스(객체) 만들기

> 인스턴스 이름 = 클래스이름()      # '>'를 이용해 인용문 표시 

__init__ 자동으로 실행되면서 새 객체가 생성된다.

```python
t1 = Toy()  # t1과 t2가 서로 다른 객체
t2 = Toy()

print(t1)
print(t2)  # 각 객체의 저장 주소를 보여줌. Toy라는 클래스에  t1과 t2라는 객체가 존재.
```

```python
t1 is t2  # False 값을 반환. 같은 클래스로 만들었지만 서로 다른 객체다!
```

# 간단한 클래스 예제

```python
class Person:
    def __init__(self, name, age, job, hobby):   #인스턴스 변수(속성) 추가
        self.name = name
        self.age = age
        self.job = job
        self. hobby =  hobby
    
    def print_info(self):
        print(f'{self.name}이(가) {self.age}살입니다!')
        print(f'{self.name}의 직업은 {self.job}입니다.')
        print(f'{self.name}의 취미는 {self.hobby}입니다.')
```

```python
saram1 = Person('honggidong', 33, 'firefighter', 'boxing')
saram2 = Person('ahnchelsu', 22, 'scientist', 'game')
saram3 = Person('kimgu', 55, 'philosopher', 'readingbooks')
saram4 = Person('leesjuil', 66, 'comedien', 'dance')
```

```python
print(saram1)
print(saram2)
print(saram3)
print(saram4)
```

```python
saram1.print_info()
```

```python
saram1.print_info()
saram2.print_info()
saram3.print_info()
saram4.print_info()
```
