# 실습 - 서울시 마포구의 cctv의 설치목적을 알아보자

```python
import json

with open('cctv.json', 'rt', encoding='utf-8') as f:
    buffer =f.read()
    cctv_list = json.loads(buffer)  # 파이썬으로 변환
    cctv_purpose = set() # 중복을 없애기 위해 결과를 담을 빈 세트를 생성, set은 중복 제거
    for place in cctv_list:
        cctv_purpose.add(place['설치목적구분']) # ser에 데이터를 추가
cctv_purpose
```
