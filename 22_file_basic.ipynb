# 파일 입출력

## 1. 파일 객체 = open( ) ----> 마지막에 파일 객체명.close( )

### 파일 열기 모드
|종류|설명|특징|
|----|----|----|
| w | 쓰기 모드(write) | 파일이 없는 경우 새로 생성, 있는 경우 덮어쓰기|
| a | 추가 모드(append) | 파일이 없는 경우 새로 생성, 있는 경우 이어서 작업|
| r | 읽기 모드(read)| 파일이 없는 경우 에러, 있는 경우 읽어서 처리|


### 파일의 종류
|옵션|파일종류|생략 유무|가능한 파일 형식|
|----|----|----|----|
|t|텍스트 파일|생략 가능|txt, csv, docx...|
|b|바이너리 파일|생략을 할 수 없다.|그림, 소리, 동영상 등|

```python

file = open('club.txt', 'wt', encoding='utf-8')
file.write('동아리 회원 목록 : 1. 라이언') 
file.close()
```

```python
f = open('club.txt', 'wt', encoding='utf-8')  # 덮어쓰기

while True:
    name= input('name : ')
    if not name:
        break
    f.write(name)
f.close()
```

```python
f = open('club.txt', 'a', encoding='utf-8')  # 추가하기

while True:
    name= input('name : ')
    if not name:
        break
    f.write(name +  '\n')
f.close()
```

```python
f =open('club.txt', 'r', encoding='utf-8')
club_members = f.read()
f.close()

print(club_members)
```

## 2. with open( ) as 파일 객체명:

```python
with open('club.txt', 'r', encoding='utf-8') as file:
    club_members = file.read()

club_members
```

```python
with open('club.txt', 'r', encoding='utf-8') as file:
    club_members = file.readlines()  # 전체 데이터를 리스트 형식으로 가져온다.

club_members
```

```python
with open('club.txt', 'r', encoding='utf-8') as file:
    club_members = file.readline()      # 맨 앞의 첫 줄 하나
club_members
```

## 실습 예제 - 오늘의 스케줄 입력하기

```python
import time

# 파일명 형식--> 2026-05-26.txt
with open(time.strftime('%Y-%m-%d') + '.txt', 'a', encoding='utf-8') as f :
    while True:
        schedule = input('오늘의 스케쥴을 입력하세요 : ')
        if not schedule:
            break
        f.write(schedule + '\n')

    
```

```python
with open('2026-05-26.txt', 'r') as f:
    schedules = f.readlines()

schedules
```

## 파일 오류 처리

- 파일이 존재하지 않는다면 오류 메세지를 출력

```python
import os.path
filename = input('filename(확장자까지 포함) : ')
if os.path.exists(filename):  #파일 경로에 파일이 존재한다면
    with open(filename, 'r', encoding='utf-8') as f:
        data = f.read()
    print(data)
else:
    print(f'{filename} 파일이 없습니다!')
```

## Python의 예외 처리

- try:
    예외(오류)가 일어날 것 같은 코드, 일어날 가망성이 있는 코드
- except:
    예외가 일어났을 때 실행되는 코드
- else:
    정상적일 때 실행되는 코드
- finally:
    예외와 상관없이 무조건 실행되는 코드

set --> try, except, else
    --> try, finally

```python
filename = input('filename(확장자까지 포함) : ')
try: #예외가 발생할 가능성이 있는 코드
    with open(filename, 'r', encoding='utf-8') as f:
        data = f.read()
    print(data)
except: # 예외가 발생하면 except 쪽으로 온다.
    print(f'{filename} 파일이 없습니다!')
```

# 번역기 만들기

```python
with open('notice.txt', 'w', encoding='utf-8') as f:
    f.write('원문 - 한국어 \n')
    f.write('2026년 6월 3일은 전국동시지방선거일입니다.\n')
    f.write('우리 강의는 그날 진행되지 않으니 참고부탁드립니다. \n')
    f.write('즐거운 하루 보내시고, 복습 꼭 하셔요! \n')
```

```python
from googletrans import Translator

translator = Translator()
with open('notice.txt', encoding='utf-8-sig') as f:
    notice =f.read()
    print('원문 - 한국어')
    print('='*80)
    print(notice)
    print('='*80)

    code = ['en', 'ja', 'zh-TW'] # 영어, 일본어, 중국어

    while True:
        choice = int(input('영어(1), 일본어(2), 중국어(3), 종료(0) : '))
        if choice == 0:
            break
        lang = code[choice-1]
        res = translator.translate(notice, src='ko', dest=lang)
        print(res.text + '\n')
```
