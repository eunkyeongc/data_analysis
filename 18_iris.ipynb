```python
# 붓꽃의 품종별 그룹화 - 평균 - 새로운 데이터 프레임으로 저장 - 차트 생성
df2 = iris.groupby(iris['species']).mean()
df2
```

```python
# 품종의 꽃받치 길이 그래프
df2.sepal_length[:].plot(kind='bar', rot=0)
plt.title('품종별 꽃받침 길이 시각화') # 차트 제목
plt.xlabel('품종') # x축 제목
plt.ylabel('꽃받침 길이의 평균값') # y축 제목
plt.show()
```

```python
# 히스토그램(빈도수 그래프)
iris.plot(kind='hist')
plt.title('각 특징 값들의 빈도수 히스토그램')
plt.show()
```

```python
# 상자 그래프 --> 이상치 발견
iris.plot(kind='box')
plt.title('각 특징 값들의 빈도수에 대한 박스 플롯')
plt.xlabel('특징') # x축 제목
plt.ylabel('데이터 값') # y축 제목
plt.show()
```

```python
# 산점도 그래프 (산포도 그래프)
iris.plot(x='sepal_length', y='petal_length', kind='scatter')
plt.show()
```

# ex) 4. 타이타닉

```python
import pandas as pd

titanic =pd.read_csv('train.csv')
titanic
```

### 1) 성별에 따른 생존율

```python
titanic2=titanic[['Sex', 'Survived']]  #성별, 생존율 
titanic2
```

```python
titanic2.groupby('Sex').mean()
```

### 2) 함께 승선한 부모와 자식수(parch)에 따른 생존율

```python
titanic3=titanic[['Parch', 'Survived']]
titanic3
```

```python
titanic3.groupby('Parch').mean()
```

```python
titanic3.groupby('Parch', as_index=False).mean()
```

### 3) 객실 등급(Pclass), 성별(Sex) 컬럼별로 생존율, 나이의 평균값

- 그룹 --> 2개
- 평균값 --> 2개

```python
titanic4=titanic[['Pclass', 'Sex', 'Survived', 'Age']]
titanic4
```

```python
titanic4.groupby(['Pclass', 'Sex']).mean()
```

#### 데이터프레임명.groupby(['계산할 컬럼1', '계산할 컬럼2'])[['계산하고 싶은 컬럼1', '계산하고 싶은 컬럼2']].mean()

```python
titanic.groupby(['Pclass', 'Sex'])[['Survived', 'Age']].mean()
```
