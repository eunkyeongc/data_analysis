# JSON
- 자바스크립트로 만들어진 경량의 데이터파일
- {"key":"value"} ---> python dictionary와 비슷하다.
- 변환이 쉽다.

```python
import json
```

- 파이썬으로 딕셔너리 --> json 변환

```python
dict_list = [
    {
        "name":'james',
        'age' : 20,
        'spec':[175.5, 70.5]
    },
    {
        "name":'alice',
        'age' : 21,
        'spec':[168.5, 60.5]
    },
    {
        "name":'danial',
        'age' : 30,
        'spec':[185.5, 90.5]
    }

]
dict_list
```

```python
type(dict_list)
```

```python
type(dict_list[2])
```

# json으로 변환

```python
json_string = json.dumps(dict_list)
```

```python
json_string
```

# json  파일로 저장(json 으로 내보내기)

```python
with open('dictlist.json', 'wt') as f:
    f.write(json_string)
```

# json 파일 읽기

```python
import json

with open('dictlist.json', 'rt') as f:
    json_reader = f.read()
    dict_list = json.loads(json_reader) # 파이썬으로 변환
dict_list
```

```python
type(dict_list)
```

```python
for dic in dict_list:
    print(f'이름 : {dic["name"]}')
    print(f'나이 : {dic["age"]}')
    print(f'키 : {dic["spec"][0]}')
    print(f'몸무게 : {dic["spec"][1]}')
    print('='*50)
```
