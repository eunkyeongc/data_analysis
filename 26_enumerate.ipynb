# 리스트의 for문

```python
pandas =['러바오', '아이바오', '푸바오', '루이바오', '후이바오']
```

```python
for p in pandas:
    print(p) # 리스트의 값(요소)이 나온다!
```

```python
for i in range(5):
    print(i)
```

# enumerate 함수

- 인덱스와 값(요소)을 함께 불러 오고싶을 때 사용하는 함수
- 튜플의 형태로 불러 옮

for 인덱스번호, 값 in enumerate(컬렉션자료형):
    pass

```python
for p in enumerate(pandas):
    print(p)
```

## 튜플 언패킹

각각의 변수에 튜플의 값을 쪼개 담는다.

```python
p[0], p[1]
```

```python
for index, name in enumerate(pandas):
    print(index)
```

```python
for index, name in enumerate(pandas):
    print(name)
```

```python
for index, name in enumerate(pandas):
    print(index, name)
```

```python
for index, name in enumerate(pandas):
    print(f'{index}번째 판다 {name}')
```

```python
for index, name in enumerate(pandas):
    print(f'{index+1}번째 판다 {name}')
```
